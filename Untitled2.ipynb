```python
import random
```

```python
import pandas as pd
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
# Вариант 1
set_lst = set(lst)
set_lst_1 = list(set_lst)
index = 0
for i in range(len(set_lst_1)):
  for j in range(len(lst)):
    if lst[j] == set_lst_1[i]:
      lst[j] = index
  index += 1
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
# Вариант 2
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
new_lst = pd.DataFrame({'robot': [1 if i == 'robot' else 0 for i in lst], 'human': [1 if j == 'human' else 0 for j in lst]})
new_lst
```
